```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("insurance.csv")
print(df.head())
```

```python
print(df.shape)
```

```python
print(df.isnull().sum())
```

```python
print(df.duplicated().sum())
```

```python
df = df.drop_duplicates()
print('Shape after dropping duplicates:', df.shape)
print('Duplicates remaining:', df.duplicated().sum())
```

```python
print('for age')
median_age = df['age'].median()
min_age = df['age'].min()
max_age = df['age'].max()
q1_age = df['age'].quantile(0.25)
q2_age = df['age'].quantile(0.5)
q3_age = df['age'].quantile(0.75)
```

```python
value_count = df['sex'].value_counts()
mode_gender = df['sex'].mode()[0]
```

```python
print(value_count)
```

```python
bmi_mean = df['bmi'].mean()
bmi_median = df['bmi'].median()
bmi_min = df['bmi'].min()
bmi_max = df['bmi'].max()
q1_bmi = df['bmi'].quantile(0.25)
q2_bmi = df['bmi'].quantile(0.5)
q3_bmi = df['bmi'].quantile(0.75)
```

```python
children_median = df['children'].median()
children_min = df['children'].min()
children_max = df['children'].max()
children_q1 = df['children'].quantile(0.25)
children_q2 = df['children'].quantile(0.5)
children_q3 = df['children'].quantile(0.75)
```

```python
smoker_value_counts = df['smoker'].value_counts()
smoker_mode = df['smoker'].mode()[0]
```

```python
region_value_counts = df['region'].value_counts()
region_mode = df['region'].mode()[0]
```

```python
charges_mean = df['charges'].mean()
charges_median = df['charges'].median()
charges_min = df['charges'].min()
charges_max = df['charges'].max()
q1_charges = df['charges'].quantile(0.25)
q2_charges = df['charges'].quantile(0.5)
q3_charges = df['charges'].quantile(0.75)
```

```python
print("DATA PROFILE")
print("Data Profile for Age:")
print(f"Median Age: {median_age}")
print(f"Min Age: {min_age}")
print(f"Max Age: {max_age}")
print(f"Q1 Age: {q1_age}")
print(f"Q2 Age: {q2_age}")
print(f"Q3 Age: {q3_age}")
print("--------------------------------------------------------------")
print("Data Profile for Gender:")
print(f"Value Counts:{value_count}")
print(f"Mode: {mode_gender}")
print("--------------------------------------------------------------")
print("Data Profile for BMI:")
print(f"Mean BMI: {bmi_mean}")
print(f"Median BMI: {bmi_median}")
print(f"Min BMI: {bmi_min}")
print(f"Max BMI: {bmi_max}")
print(f"Q1 BMI: {q1_bmi}")
print(f"Q2 BMI: {q2_bmi}")
print(f"Q3 BMI: {q3_bmi}")
print("--------------------------------------------------------------")
print("Data Profile for Children:")
print(f"Median Children: {children_median}")
print(f"Min Children: {children_min}")
print(f"Max Children: {children_max}")
print(f"Q1 Children: {children_q1}")
print(f"Q2 Children: {children_q2}")
print(f"Q3 Children: {children_q3}")
print("--------------------------------------------------------------")
print("Data Profile for Smoker:")
print(f'Value Counts:\n{smoker_value_counts}')
print(f'Mode: {smoker_mode}')
print("--------------------------------------------------------------")
print("Data Profile for Region:")
print(f'Value Counts:\n{region_value_counts}')
print(f'Mode: {region_mode}')
print("--------------------------------------------------------------")
print("Data Profile for Charges:")
print(f'Mean Charges: {charges_mean}')
print(f'Median Charges: {charges_median}')
print(f'Min Charges: {charges_min}')
print(f'Max Charges: {charges_max}')
print(f'Q1 Charges: {q1_charges}')
print(f'Q2 Charges: {q2_charges}')
print(f'Q3 Charges: {q3_charges}')
print("--------------------------------------------------------------")
```
